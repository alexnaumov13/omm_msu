\psi_{n, m}^{j+\frac{1}{2}}=\left.\left(\Delta u-\frac{\partial u}{\partial t}+f\right)\right|_{\left(x_{n}, y_{m}, t_{j+\frac{1}{2}}\right)}+O\left(h_{x}^{2}\right)+O\left(h_{y}^{2}\right)+O\left(\tau^{2}\right)= \bigg[\left.\left(\Delta u-\frac{\partial u}{\partial t}+f\right)\right|_{\left(x_{n}, y_{m}, t_{j+\frac{1}{2}}\right)} = 0 \bigg]=O\left(h_{x}^{2}\right)+O\left(h_{y}^{2}\right)+O\left(\tau^{2}\right)
$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å—Ö–µ–º–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–π –∏–º–µ–µ—Ç –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏
$O\left(h_{x}^{2}\right)+O\left(h_{y}^{2}\right)+O\left(\tau^{2}\right),$ –∞ –∑–Ω–∞—á–∏—Ç, –≤ —Å–∏–ª—É –ª–∏–Ω–µ–π–Ω–æ—Å—Ç–∏ –∏ –±–µ–∑—É—Å–ª–æ–≤–Ω–æ–π —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏, –æ–Ω–∞ —Å—Ö–æ-
–¥–∏—Ç—Å—è –∏ –∏–º–µ–µ—Ç –≤—Ç–æ—Ä–æ–π –ø–æ—Ä—è–¥–æ–∫ —Ç–æ—á–Ω–æ—Å—Ç–∏ –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º –∏ –≤—Ä–µ–º–µ–Ω–∏. –¢–∞–∫ –∫–∞–∫ –¥–ª—è –µ–µ pea-
–ª–∏–∑–∞—Ü–∏–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è —á–∏—Å–ª–æ –¥–µ–π—Å—Ç–≤–∏–π, –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª—É —É–∑–ª–æ–≤ —Å–µ—Ç–∫–∏, —Å—Ö–µ–º–∞ —è–≤–ª—è–µ—Ç—Å—è
—ç–∫–æ–Ω–æ–º–∏—á–Ω–æ–π.

# –ü—Ä–æ–≥—Ä–∞–º–Ω–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è

–í–≤–µ–¥–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

–í–≤–µ–¥–µ–º –æ–±–ª–∞—Å—Ç—å —Ä–µ—à–µ–Ω–∏—è, —á–∏—Å–ª–æ —É–∑–ª–æ–≤ —Å–µ—Ç–∫–∏, –Ω–∞—á–∞–ª—å–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ , —à–∞–≥–∏ $ h_x, h_y, \tau$ –∏ —Å–∞–º—É —Å–µ—Ç–∫—É

```python
X_min = 0. 
X_max = 2.

Y_min = 0.
Y_max = 1.

T_min = 0. 
T_max = 0.2

#—á–∏—Å–ª–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
N = 50
M = 50
J = 50

#—Å–µ—Ç–∫–∞ –∏ —à–∞–≥–∏
# –°–µ—Ç–∫–∞ –ø–æ x, y –∏ t
x = np.linspace(X_min, X_max , N + 1)
y = np.linspace(Y_min , Y_max, M + 1)
t = np.linspace(T_min , T_max, J + 1)

# –®–∞–≥ –ø–æ x, y –∏ t
hx = (X_max - X_min)/float(N)
hy = (Y_max - Y_min  ) / float(M)
tau  = (T_max-T_min) / float(J)

print hx

def u_init(x,y): 
    return math.cos(math.pi*x/4)*y*(1-y)

# –ù–µ–æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç—å
def f(x,y,t):
    return math.pow(y,2)*math.pow(t,2)
```

–î–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–º —Ñ—É–Ω–∫—Ü–∏–∏, –∑–∞–ø–æ–ª–Ω—è—é—â—É—é –º–∞—Å—Å–∏–≤—ã $A, B, C, F$ –ø—Ä–∏ –ø–µ—Ä–µ—Ö–æ–¥–µ:$\\$

1)–æ—Ç –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Å–ª–æ—è $j\rightarrow j+\frac{1}{2}$

2)–æ—Ç –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Å–ª–æ—è $j+\frac{1}{2} \rightarrow j+1$

```python
def Set_diag_1(f, N, M, J, x, y, t, hx, hy, tau, W, A, B, C, F, m, j):
    A[0] =  0
    B[0] = -2./(3.*math.pow(hx, 2)) - 1./(0.5*tau)
    C[0] = 2./(3.*math.pow(hx, 2))
    F[0] =  - W[j][1][m]/(0.5*tau) - (W[j][1][m+1] - 2.*W[j][1][m] + W[j][1][m-1])/math.pow(hy, 2) - f(x[1], y[m], (t[j]+t[j+1])/2)
    for n in range(2, N - 1):
        A[n-1] =  1./math.pow(hx, 2)
        B[n-1] = -2./math.pow(hx, 2) - 1/(0.5 * tau)
        C[n-1] =  1./math.pow(hx, 2)
        F[n-1] =  - W[j][n][m]/(0.5 * tau) - (W[j][n][m+1] - 2*W[j][n][m] + W[j][n][m-1])/math.pow(hy, 2) - f(x[n], y[m], (t[j]+t[j+1])/2)
    A[N-2] = 1 / math.pow(hx, 2)
    B[N-2] = -2./math.pow(hx, 2) - 1./(0.5*tau)
    C[N-2] =  0.
    F[N-2] =  - W[j][N-1][m]/(0.5*tau) - (W[j][N-1][m+1] - 2.*W[j][N-1][m] + W[j][N-1][m-1])/math.pow(hy, 2) - f(x[N-1], y[m], (t[j]+t[j+1])/2)
```

```python
def Set_diag_2( f, N, M, J, x, y, t, hx, hy, tau, V, A, B, C, F, n, j):
    A[0] =   0.
    B[0] =  -2./ math.pow(hy, 2) - 1/(0.5*tau) 
    C[0] =   1./ math.pow(hy, 2) 
    F[0] =  - V[n][1]/(0.5*tau) - (V[n+1][1] - 2*V[n][1] + V[n-1][1])/math.pow(hx, 2) - f(x[n], y[1], (t[j]+t[j+1])/2)
    for m in range(2, M - 1):
        A[m-1] =    1. / math.pow(hy, 2)
        B[m-1] =   -2. / math.pow(hy, 2) - 1/(0.5*tau)
        C[m-1] =    1. / math.pow(hy, 2)
        F[m-1] =  - V[n][m]/(0.5*tau) - (V[n+1][m] - 2*V[n][m] + V[n-1][m])/math.pow(hx, 2) - f(x[n], y[m], (t[j]+t[j+1])/2)
    A[M-2] =    1. / math.pow(hy, 2) 
    B[M-2] =   -2. / math.pow(hy, 2) - 1/(0.5*tau) 
    C[M-2] =    0.
    F[M-2] =  - V[n][M-1]/(0.5*tau) - (V[n+1][M-1] - 2.*V[n][M - 1] + V[n-1][M-1])/math.pow(hx, 2) - f(x[n], y[1], (t[j]+t[j+1])/2)
```

–¢–∞–∫–∂–µ –≤–≤–µ–¥–µ–º —Ñ—É–Ω–∫—Ü–∏—é –ø—Ä–æ–≥–æ–Ω–∫–∏

```python
def progonka(n, A, B, C, F):

    X = np.zeros(n)

    for i in range(1,n):
        m = A[i] / B[i - 1]
        B[i] = B[i] - m * C[i - 1]
        F[i] = F[i] - m * F[i - 1]

    X[n - 1] = F[n - 1] / B[n - 1]

    for i in range(n-2, -1, -1):
        X[i] = (F[i] - C[i] * X[i + 1]) / B[i]

    return X
```

–†–µ–∞–ª–∏–∑—É–µ–º —á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ

```python
# –ú–∞—Ç—Ä–∏—Ü–∞ —Å–µ—Ç–æ—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
W = np.zeros((J + 1, N + 1, M+ 1))
# –ú–∞—Ç—Ä–∏—Ü–∞ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–æ–≥–æ k+1/2-–≥–æ –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Å–ª–æ—è
V = np.zeros((N + 1, M + 1))

# –ó–∞–ø–∏—à–µ–º –Ω–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è
for n in range(0, N + 1):
    for m in range(0, M + 1):
        W[0][n][m] = u_init(x[n], y[m])
        


#–ó–∞–ø—É—Å—Ç–∏–º —Ü–∏–∫–ª –ø–æ –≤—Å–µ–º –≤—Ä–µ–º–µ–Ω–Ω—ã–º —Å–ª–æ—è–º
for j in range(0,J):
        # –ü–æ—Å—Ç—Ä–æ–∏–º k+1/2 –≤—Ä–µ–º–µ–Ω–Ω–æ–π —Å–ª–æ–π
        
        A = np.zeros(N - 1) # –ù–∏–∂–¥–Ω—è—è –∫–æ–¥–∏–∞–≥–æ–Ω–∞–ª—å
        B = np.zeros(N - 1) # –ì–ª–∞–≤–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å
        C = np.zeros(N - 1) # –í–µ—Ä—Ö–Ω—è—è –∫–æ–¥–∏–∞–≥–æ–Ω–∞–ª—å
        F = np.zeros(N - 1) # –ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å
        
        # –ü—Ä–∏ –∫–∞–∂–¥–æ–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º m –æ—Ç 1 –¥–æ M
        for m in range(1, M):
            
            # –ó–∞–¥–∞–Ω–∏–µ –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ —É —Ç—Ä—ë—Ö–¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
            Set_diag_1( f,N, M, J, x, y, t, hx, hy, tau, W, A, B, C, F, m, j)
            
            # –ú–µ—Ç–æ–¥ –ø—Ä–æ–≥–æ–Ω–∫–∏
            X = progonka(N - 1, A, B, C, F)
            
            # –ó–∞–ø–∏—Å—å —Ä–µ—à–µ–Ω–∏—è –≤ —Å–µ—Ç–∫—É
            for n in range(1, N): 
                V[n][m] = X[n-1]
                
            # –ó–∞–ø–∏—Å—å –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π –≤ —Å–µ—Ç–∫—É
            V[0][m] = (4.*V[1][m] - V[2][m])/3.
            V[N][m] = 0.
        
        
        # –ü–æ—Å—Ç—Ä–æ–∏–º k+1 –≤—Ä–µ–º–µ–Ω–Ω–æ–π —Å–ª–æ–π

        A = np.zeros(M - 1)  # –ù–∏–∂–¥–Ω—è—è –∫–æ–¥–∏–∞–≥–æ–Ω–∞–ª—å
        B = np.zeros(M - 1)  # –ì–ª–∞–≤–Ω–∞—è –¥–∏–∞–≥–æ–Ω–∞–ª—å
        C = np.zeros(M - 1)  # –í–µ—Ä—Ö–Ω—è—è –∫–æ–¥–∏–∞–≥–æ–Ω–∞–ª—å
        F = np.zeros(M - 1)  # –ü—Ä–∞–≤–∞—è —á–∞—Å—Ç—å
        
        # –ü—Ä–∏ –∫–∞–∂–¥–æ–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º n –æ—Ç 1 –¥–æ N
        for n in range(1, N):
            
            # –ó–∞–¥–∞–Ω–∏–µ –¥–∏–∞–≥–æ–Ω–∞–ª–µ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ —É —Ç—Ä—ë—Ö–¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
            Set_diag_2(f,N, M, J, x, y, t, hx, hy, tau, V, A, B, C, F, n, j)
            
            # –ú–µ—Ç–æ–¥ –ø—Ä–æ–≥–æ–Ω–∫–∏
            Y = progonka(M - 1, A, B, C, F)
            
            # –ó–∞–ø–∏—Å—å —Ä–µ—à–µ–Ω–∏—è –≤ —Å–µ—Ç–∫—É
            for m in range(1, M): 
                W[j+1][n][m] = Y[m-1]
            
            # –ó–∞–ø–∏—Å—å –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π –≤ —Å–µ—Ç–∫—É
            W[j+1][n][0] = 0.
            W[j+1][n][M] = 0.
                
                
        # –°–Ω–æ–≤–∞ —É—á—Ç—ë–º –≥—Ä–∞–Ω–∏—á–Ω—ã–µ —É—Å–ª–æ–≤–∏—è –ø–æ x, –Ω–æ —É–∂–µ –Ω–∞ —Å–ª–æ–µ j+1
        for m in range(1, M+1):
            W[j+1][0][m] = (4*W[j+1][1][m] - W[j+1][2][m])/3.
            W[j+1][N][m] = 0.
            
```

–ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —Ä–µ—à–µ–Ω–∏—è –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö $t = 0$
